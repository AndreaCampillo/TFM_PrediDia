# EDA 2021_2022

```python
url21_22 = 'https://github.com/AndreaCampillo/TFM_PrediDia/raw/Datasets/2021_22DataSet_Diabeticos_NoDiabeticos_Depurado.csv'
diabetes_df_2021_22 = pd.read_csv(url21_22, sep=';')
```

```python
print('Numero of filas: ', len(diabetes_df_2021_22) , ' y columnas: ', len(diabetes_df_2021_22.columns) , '\n')
```

```python
diabetes_df_2021_22.info()
```

```python
#Blancos
blancos_2021_22 = diabetes_df_2021_22.isnull().sum()
print(blancos_2021_22)
```

```python
diabetes_df_2021_22.head()
```

```python
#Número de registros por categoría
frecuencias = {}

for col in diabetes_df_2021_22.columns:
    frecuencias[col] = diabetes_df_2021_22[col].value_counts()

# Mostrar las frecuencias por columna
for col, frecuencias in frecuencias.items():
    print("Columna:", col)
    print(frecuencias)
    print()
```

```python
#Nombres de las columnas
column_names2021_22 = diabetes_df_2021_22.columns.tolist()
print(column_names2021_22)
```

```python
#Correlaciones lineales Gr_diabetes con todo
df_atributos_Gr_diabetes_2021_22 = diabetes_df_2021_22[['CatBMI', 'Stroke', 'HeartDis', 'PhysExer', 'HealthIns', 'NoMedCost', 'GenHealth', 'CogDiff', 'Depression', 'MentalHlth', 'MentalState', 'PhysHlth', 'WalkDiff', 'Gender', 'AgeRange', 'EdLevel', 'AnnIncome', 'SocClass', 'UrologyDz', 'VisionDiff', 'Asthma', 'LungDiseases', 'Arthritis', 'SmokerTrad', 'ECigSmok', 'AlcDrinker', 'Race', 'MaritalSt', 'LastMedChk', 'Awareness', 'FootIrrita', 'FecFootIrrita', 'HighBP', 'HighChol', 'FruitCons', 'VegCons', 'FruitOrVegCon', 'FruitAndVegCon', 'SleepHours', 'MarijuanaCon', 'GrDiabetes']]
correlaciones_Gr_diabetes_2021_22 = df_atributos_Gr_diabetes_2021_22.corr()
columna_Gr_diabetes_2021_22 = correlaciones_Gr_diabetes_2021_22['GrDiabetes'].sort_values(ascending=False)
print(columna_Gr_diabetes_2021_22)
#Lógico que Presión y colesterol den menores porque en 2022 no hay datos
```
